**Descriptive Analytics for Numerical Colummns**

```python
#Import libraries
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#Read the file
df = pd.read_csv('sales_data_with_discounts.csv')
```

```python
#find the numerical column
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()
print(numerical_columns)
```

```python
#perform Statistics
statistics = {}
for column in numerical_columns:
    column_stats = {
        'mean': df[column].mean(),
        'median': df[column].median(),
        'mode': df[column].mode()[0],
        'std_dev': df[column].std()
    }
    statistics[column] = column_stats

#Print statistics
for column, stats in statistics.items():
    print(f"Statistics for {column}:")
    print(f"Mean: {stats['mean']}")
    print(f"Median: {stats['median']}")
    print(f"Mode: {stats['mode']}")
    print(f"Standard Deviation: {stats['std_dev']}")
    print("\n")

```

**DATA VISUALIZATION**
     

**HISTOGRAM**

```python
# Import libraries for plotting
import matplotlib.pyplot as plt
import seaborn as sns

#Plot histograms for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    #Calculate skewness
    skewness = df[column].skew()

    #outliers using 1.5*IQR rule
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    #Print analysis
    print(f'Analysis for {column}:')
    print(f'Skewness: {skewness}')
    print(f'Number of outliers: {len(outliers)}')
    print('Outliers:')
    print(outliers[column].values)
    print('\n')


    print(f'Inference for {column}:')
    if skewness > 0:
        print('The distribution is positively skewed (right-skewed).')
    elif skewness < 0:
        print('The distribution is negatively skewed (left-skewed).')
    else:
        print('The distribution is approximately symmetric.')

    if len(outliers) > 0:
        print(f'There are {len(outliers)} outliers detected.')
    else:
        print('No significant outliers detected.')
    print('\n')
```

**BOX PLOT**

```python
#boxplots for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.show()

    #interquartile range (IQR)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Identify outliers using 1.5*IQR rule
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    # Print findings
    print(f'Findings for {column}:')
    print(f'Interquartile Range (IQR): {IQR}')
    print(f'Lower Bound: {lower_bound}')
    print(f'Upper Bound: {upper_bound}')
    print(f'Number of outliers: {len(outliers)}')
    if len(outliers) > 0:
        print('Outliers:')
        print(outliers[column].values)
    print('\n')

```

**BAR CHART**

```python
#Identify categorical columns
categorical_columns = df.select_dtypes(include=[object, 'category']).columns.tolist()

#bar charts for each categorical column
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=df[column], order=df[column].value_counts().index)
    plt.title(f'Bar Chart of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

    # Analyze distribution
    value_counts = df[column].value_counts()
    print(f'Analysis for {column}:')
    print(value_counts)
    print('\n')
```

**Standardization of Numerical Variables**

```python
#Standardize numerical columns
standardized_df = df.copy()
for column in numerical_columns:
    mu = df[column].mean()
    sigma = df[column].std()
    standardized_df[column] = (df[column] - mu) / sigma

#Show before and after comparisons of the data distributions
for column in numerical_columns:
    plt.figure(figsize=(14, 6))

    # Before standardization
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Before Standardization: {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # After standardization
    plt.subplot(1, 2, 2)
    sns.histplot(standardized_df[column], kde=True)
    plt.title(f'After Standardization: {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    plt.show()
```

**Conversion of Categorical Data into Dummy Variables**

```python
#Identify categorical columns
categorical_columns = df.select_dtypes(include=[object, 'category']).columns.tolist()

# Apply encoding to categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

#  Display a portion of the transformed dataset
print(df_encoded.head())
```
